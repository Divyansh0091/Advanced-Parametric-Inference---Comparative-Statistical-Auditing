```python
import pandas as pd 
import numpy as np
import scipy.stats as stat
data = {
    'Employee_ID':[101,102,103,104,105,106,107,108,109,110],
    'Before_Automation':[20,30,22,28,35,27,36,40,72,45],
    'After_Automation':[30,45,65,36,40,26,60,65,70,80]
}
df = pd.DataFrame(data)
df
    
```

```python
t_test,p_value = stat.ttest_rel(df['Before_Automation'],df['After_Automation'])
print(f"Statistic-Test{t_test}") 
print(f"P-Value{p_value}")
alpha = 0.05
if p_value < alpha:
    print("Result: Hypothesis test Rejected")
else:
    print("Result: Hypthesis test not Rejected")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_long = df.melt(id_vars='Employee_ID', value_vars=['Before_Automation', 'After_Automation'], var_name='Automation', value_name='Performance')

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.boxenplot(x='Automation', y='Performance', data=df_long, palette='pastel')

plt.title('Impact Analysis: Pre vs Post Automation Strategy', fontsize=14, fontweight='bold')
plt.xlabel('Implementation Stage', fontsize=12)
plt.ylabel('Assessment Score', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
mean1,std1,nobs1 = 25,4,30
mean2,std2,nobs2 = 22,5,35
t_test,p_value = stat.ttest_ind_from_stats(mean1,std1,nobs1,mean2,std2,nobs2)
print(f"Statistic-test{t_test}") 
print(f"P-value{p_value}") 
alpha = 0.05 
if p_value < alpha:
    print(f"Result: Hypothesis Rejected") 
else:
    print(f"Result: Hypothesis not Rejected") 
```
